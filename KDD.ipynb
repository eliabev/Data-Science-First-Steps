## Matriz de Correlação de Pearson Para Variáveis
O valor 1 indica uma correlação direta perfeita, i.e., os dados andam sempre alinhados; onde um cresce, assim o outro também.
O valor "-1" indica correlação negativa perfeita; onde um cresce, o outro dado diminui

```python
mask = np.triu(np.ones_like(dados.corr(), dtype=np.bool))
plt.figure(figsize=(20, 20))
# sns.heatmap(dados.corr(), mask=mask, square=True, annot=True, vmin=-1, vmax=1)
sns.scatterplot(data=dados)
plt.show()
```
