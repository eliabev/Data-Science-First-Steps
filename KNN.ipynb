# KNN Com Dataset Iris Flower

```python
from sklearn import datasets
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
iris.target_names
```

```python
df['target_name'] = iris.target_names[df['target']]
df
```

```python
iris_features=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
y = df.target
X = df[iris_features]
y
```

```python
X
```

# Usar Scikit-Learn

**Define**: Escolha de modelo (define parâmetros)

**Fit**: Treinar

**Predict**: Fazer predição

**Evaluate**: Avaliar resultados

```python
modelo = KNeighborsClassifier(17) # isso significa que "k" é igual a 3
modelo.fit(X, y)
X
```

```python
modelo.predict(X) # treinando e testando com o mesmo conjunto o bicho ainda consegue errar. só por Deus mesmo
```

Verificamos a acurácia do modelo gerado, em porcentagem

```python
score_final = modelo.score(X, y) * 100
print(f'Acurácia do algoritmo: {score_final:.2f}%')
```
