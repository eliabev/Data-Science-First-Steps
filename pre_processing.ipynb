```python
import pandas as pd
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv')
dados.head(5)
```

```python
dados.dropna(subset=['loser_rank'], inplace=True) # Apaga a linha quando subset 'loser_rank' for null
```

```python
dados
```

```python
dados.dropna(axis=1, how='all') # apaga a coluna em que todos os valores são null
```

```python
dados['loser_entry'].unique()
```

```python
dados['loser_entry'].fillna(value='X', inplace=True) # substitui os valores null da coluna loser_entry por X
dados['loser_entry'].unique()
```

```python
dados['w_ace'].unique()
```

```python
dados.fillna(value={'w_ace': 0, 'loser_ht': dados['loser_ht'].mean()}, inplace=True)
dados.w_ace.unique()
```

```python
dados.loser_ht.unique()
```

```python
dados.minutes[60:110]
```

```python
dados.minutes.fillna(method='ffill', inplace=True)
dados.minutes.isnull()
```

```python
dados.duplicated()
```

```python
dados.duplicated().sum()
```

```python
data2 = dados.drop_duplicates(subset=['tourney_name']) # sem o 'inplace=True' as mudanças não são geradas dentro do próprio conjunto, o método apenas retorna o conjunto resultante
data2
```
